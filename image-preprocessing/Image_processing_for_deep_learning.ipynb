# Grayscale Images

Grayscale images are images that have a single color channel. This means that the value of each pixel represents the amount of light, or intensity, that particular pixel carries. Grayscale images are also called black-and-white, although there are many shades of gray.

if a image is of dim 200* 200 - we have a matrix of 200*200

0 - black
255 - white
0-255 - represents white color intensity

# RGB Image

RGB images are images that have three color channels. This means that each pixel carries three values that represent the intensity of the red, green, and blue color channels. The combination of these three values for each pixel can produce almost any color. RGB images are also called true color images because they can produce the full spectrum of colors.

if a image is of dim 200* 200 - we have a matrix of 200*200*3 - have 3 matrices
R = Red - 0 - no red , 255 - completely red
G = Green
B = Blue

255,255,255 = white
0, 0, 0 = black
